```python
#since there are 67 nan values for salary we will replace it with 0 
#nan values in this column signify that candidate has not been placed
mba_df['salary']=mba_df['salary'].fillna(0)
```

```python
#drop unnecessary columns like sl_no, ssc_b (ssc board),hsc_b (hsc board)
mba_df=mba_df.drop(columns=['sl_no','ssc_b','hsc_b','salary'])
mba_df.head(5)
```

```python
# use label encoding for columns which have only 2 possible values
from sklearn.preprocessing import LabelEncoder

#create instance of label encoder
labelencoder=LabelEncoder()

#assign values and create new column
mba_df['gender']=labelencoder.fit_transform(mba_df['gender'])
mba_df['workex']=labelencoder.fit_transform(mba_df['workex'])
mba_df['specialisation']=labelencoder.fit_transform(mba_df['specialisation'])
mba_df['status']=labelencoder.fit_transform(mba_df['status'])
mba_df
```

```python
# generate binary values using get_dummies
hsc_df = pd.get_dummies(mba_df, columns=["hsc_s"], prefix=["hsc_stream"] )
mba_ml_df = pd.get_dummies(hsc_df, columns=["degree_t"], prefix=["degree"] )
mba_ml_df.head(5)
```
